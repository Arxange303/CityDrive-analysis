# Download data

# 1) how much income does Qashqai bring in a month in Moscow?

## therefore, I take into account only cases when information is in the table with a profit

```python
# Б)I summarize the duplicates in expenses. Example:
df_fuel_cost[df_fuel_cost.duplicated(['car_id', 'month'])]
```

```python
df_fuel_cost_new = df_fuel_cost.groupby(['car_id','month']).sum().reset_index()
```

```python
# Collecting everything in one table
a = df.merge(df_leasing_payments, how = 'left',left_on='car_id_l',right_on='car_id')
b = a.merge(df_fuel_cost_new,  how = 'left',left_on=['car_id_l','month'],right_on=['car_id','month'])
c = b.merge(df_maintenance_cost,  how = 'left',left_on=['car_id_l','month'],right_on=['car_id','month'])
c.head(2)
```

```python
# Less all expenses from income. Most profitable cars/cities
c['earning'] = c['profit']-c['fuel_cost'] - c['maintenance_cost']
c.groupby(['model','city']).earning.median().to_frame().sort_values(by=['earning'],ascending=False).head(3)
d = c.groupby(['model','city']).earning.median().to_frame().reset_index()
d['row_num'] = d.sort_values([ 'earning'], ascending=False).groupby([ 'city']).cumcount() + 1
d[d['row_num']==1]
```

```python
import matplotlib.pyplot as plt
for i in d['city'].unique():
    d1=d[d['city']==i]
    fig, ax = plt.subplots(figsize=(6, 6))
    labels=d1['model']
    patches, texts, pcts = ax.pie(
        d1['earning'], labels=labels, autopct='%.1f%%',
        wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
        textprops={'size': 'x-large'},
        startangle=90)
    # For each wedge, set the corresponding text label color to the wedge's
    # face color.
    for j, patch in enumerate(patches):
      texts[j].set_color(patch.get_facecolor())
    plt.setp(pcts, color='white')
    plt.setp(texts, fontweight=600)
    ax.set_title(f'Distribution of income by models in{i}', fontsize=18)
    plt.tight_layout()

```
